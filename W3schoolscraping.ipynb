```python
import requests
from bs4 import BeautifulSoup
```

```python
url="https://www.w3schools.com/"
content = requests.get(url)
page=content.content
soup = BeautifulSoup(page,'html5lib')
soup
```

```python
name=[]
links=[]
for data in soup.find_all('div', {"class" : "w3-col l3 m6"}):
    for link in data.find_all('a'):
        #print(link)
        names = link.contents[0]
        fullLink = link.get('href')
        #print(names)
        #print("https://www.w3schools.com/"+fullLink)
        name.append(names)
        links.append("https://www.w3schools.com/"+fullLink)
print(name,links)
```

```python
import pandas as pd     
dict = {'Name': name, 'Links': links}  
df = pd.DataFrame(dict)     
df.to_csv('W3schooltask.csv' , index=False)
```
